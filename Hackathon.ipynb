```python
#Started with simple XGBoost

import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix, roc_curve
from sklearn.preprocessing import StandardScaler


X = df.drop(columns=['ID', 'target'])  
y = df['target'] 

# Splitting data into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Handle class imbalance
scale_pos_weight = y_train.value_counts()[0] / y_train.value_counts()[1]


scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)

# XGBoost model setup
xgb_model = xgb.XGBClassifier(
    objective='binary:logistic', 
    scale_pos_weight=scale_pos_weight,  
    eval_metric='auc', 
    use_label_encoder=False,  
    random_state=42
)

# Training the model
xgb_model.fit(X_train, y_train)

# Predicting on validation set
y_pred = xgb_model.predict(X_val)
y_pred_proba = xgb_model.predict_proba(X_val)[:, 1]  # Probability for the positive class
```

```python
# Model evaluation
print("Classification Report:")
print(classification_report(y_val, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_val, y_pred))

# AUC-ROC Score
roc_auc = roc_auc_score(y_val, y_pred_proba)
print(f"AUC-ROC: {roc_auc}")

# Plotting the ROC Curve
fpr, tpr, thresholds = roc_curve(y_val, y_pred_proba)
plt.figure()
plt.plot(fpr, tpr, color='blue', label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

Tried Grid search but its taking forever. So moved to Randomsearch huperparameter tuning, Reduced  model fits in cross validation and also brought in parallel processing (All to reduce time and better performance)

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix, roc_curve
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


X = df.drop(columns=['ID', 'target'])  
y = df['target']  

# Splitting data into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Handle class imbalance
scale_pos_weight = y_train.value_counts()[0] / y_train.value_counts()[1]


scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)

# XGBoost model setup
xgb_model = xgb.XGBClassifier(
    objective='binary:logistic', 
    scale_pos_weight=scale_pos_weight,  # Handling class imbalance
    eval_metric='auc', 
    use_label_encoder=False,  # Avoids a warning
    random_state=42
)

# Defining the parameter distribution for RandomizedSearchCV
param_dist = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'min_child_weight': [1, 3, 5],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'n_estimators': [100, 200, 300],
    'gamma': [0, 0.1, 0.5],
    'reg_alpha': [0, 0.01, 0.1],
    'reg_lambda': [1, 1.5, 2]
}

# Setting up RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist,
    n_iter=100,  
    scoring='roc_auc',
    cv=3,  # 3-fold cross-validation to reduce time
    verbose=1,
    random_state=42,
    n_jobs=-1  # Use all available cores
)

# Fitting the model
random_search.fit(X_train, y_train)

# Output the best parameters found by RandomizedSearchCV
print("Best Parameters:", random_search.best_params_)

# Predicting on the validation set with the best model
best_model = random_search.best_estimator_
y_pred = best_model.predict(X_val)
y_pred_proba = best_model.predict_proba(X_val)[:, 1]  # Probability for the positive class

# Model evaluation
print("Classification Report:")
print(classification_report(y_val, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_val, y_pred))

# AUC-ROC Score
roc_auc = roc_auc_score(y_val, y_pred_proba)
print(f"AUC-ROC: {roc_auc}")

# Plotting the ROC Curve
fpr, tpr, thresholds = roc_curve(y_val, y_pred_proba)
plt.figure()
plt.plot(fpr, tpr, color='blue', label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

The high threshold (0.9998) means the model is only predicting class 1 when it is extremely confident, which is why the precision is high but recall is very low.

So, this was modified to find the best threshold based on the highest F1 score.

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix, roc_curve, precision_recall_curve, f1_score, average_precision_score
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt

# Best hyperparameters from RandomizedSearchCV
best_params = {
    'subsample': 1.0,
    'reg_lambda': 1.5,
    'reg_alpha': 0.01,
    'n_estimators': 300,
    'min_child_weight': 5,
    'max_depth': 7,
    'learning_rate': 0.1,
    'gamma': 0,
    'colsample_bytree': 0.8,
    'n_jobs': -1
}


X = df.drop(columns=['ID', 'target'])  # Features
y = df['target']  # Target

# Splitting data into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Handle class imbalance
scale_pos_weight = y_train.value_counts()[0] / y_train.value_counts()[1]


scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)

# XGBoost model setup with best hyperparameters and adjusted class weights
xgb_model = xgb.XGBClassifier(
    objective='binary:logistic',
    scale_pos_weight=scale_pos_weight,  # Handling class imbalance
    eval_metric='auc',
    use_label_encoder=False,
    random_state=42,
    **best_params  # Use best hyperparameters from RandomizedSearchCV
)

# Training the model
xgb_model.fit(X_train, y_train)

# Predicting on the validation set
y_pred_proba = xgb_model.predict_proba(X_val)[:, 1]  # Probability for the positive class

# Compute precision-recall pairs for different probability thresholds
precision, recall, thresholds = precision_recall_curve(y_val, y_pred_proba)

# Find the threshold with the best F1 score
best_f1 = 0
best_threshold = 0.5

for threshold in thresholds:
    y_pred_temp = (y_pred_proba > threshold).astype(int)
    f1 = f1_score(y_val, y_pred_temp)
    if f1 > best_f1:
        best_f1 = f1
        best_threshold = threshold

print(f"Best Threshold for F1 Score: {best_threshold}")

# Apply the best threshold to predictions
y_pred_adjusted = (y_pred_proba > best_threshold).astype(int)

# Model evaluation
print("Classification Report:")
print(classification_report(y_val, y_pred_adjusted))

print("Confusion Matrix:")
print(confusion_matrix(y_val, y_pred_adjusted))

# AUC-ROC Score
roc_auc = roc_auc_score(y_val, y_pred_proba)
print(f"AUC-ROC: {roc_auc}")

# Average Precision Score
ap_score = average_precision_score(y_val, y_pred_proba)
print(f"Average Precision Score: {ap_score}")

# Plotting the ROC Curve
fpr, tpr, _ = roc_curve(y_val, y_pred_proba)
plt.figure()
plt.plot(fpr, tpr, color='blue', label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```
